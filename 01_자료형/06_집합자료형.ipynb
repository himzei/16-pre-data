```python
# 집합자료형은 어떻게 만들까? 
s1 = set([1, 2, 3])
s2 = {4, 5, 6}

print(s1)
print(type(s1))
print(s2)
print(type(s2))
```

```python
# 집합자료형의 특징
## 1. 중복을 허용하지 않습니다. 
## 2. 순서가 없다 
```

```python
# 교집합 구하기
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
print(s1)
print(s2)

## intersection()
print(s1.intersection(s2))

## &
print(s1 & s2)
```

```python
# 합집합 구하기
## union()
print(s1.union(s2))

## |
print(s1 | s2)
```

```python
# 차집합 구하기 
## difference()
## s1 - s2
print(s1.difference(s2))
print(s1 - s2)
## s2 - s1
print(s2.difference(s1))
print(s2 - s1)
```

```python
# 집합자료형 관련 함수 
## 값 1개 추가하기 - add 

s1 = {1, 2, 3}
s1.add(4)

print(s1)
```

```python
## 값 여러개 추가하기 - update
s1 = {1, 2, 3}
s1.update([4, 5, 6])
print(s1)
```

```python
## 특정 값 제거하기 - remove
s1 = {1, 2, 3}
s1.remove(2)
s1
```

```python
## 모든값 제거하기 
s1 = {1, 2, 3}
s1.clear()
s1
```
