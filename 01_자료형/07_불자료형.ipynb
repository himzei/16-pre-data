```python
# if 0: 
#     수행할 문장 

print(bool("python"))
print(bool(""))
print(bool(" "))

print(bool(10))
print(bool(0))

print(bool([1, 2, 3]))
print(bool([]))

print(bool({"name": "pey"}))
print(bool({}))
```

```python
# 논리연산자 

## and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
# or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
# not 연산자
print(not True)
```
