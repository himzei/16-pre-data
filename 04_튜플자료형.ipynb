```python
# 튜플은 어떻게 만들까? (tuple)
t1 = ()
t2 = (1, )
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ('a', 'b', ('ab', 'cd'))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))
print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```

```python
a = 1,
print(type(a))
```

```python
# 튜플의 요솟값을 지우거나 변경하려고 하면 어떻게 될까? 

t1 = (1, 2, 'a', 'b')
del t1[0]
```

```python
# 튜플 요소값을 변경하려 할 때
t1 = (1, 2, 'a', 'b')
t1[0] = 'c'
```

```python
# 튜플 인덱싱
t1 = (1, 2, 'a', 'b')
print(t1[2])
print(t1[-1])
```

```python
# 튜플슬라이싱
t1 = (1, 2, 'a', 'b')
t1[1:]
```

```python
# 튜플 더하기 
t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
print(t1 + t2)
print(t1)
```

```python
t2 = (3, 4)
print(t2 * 3)
print(t2)
```
