```python
#1. 일반적인 함수
## 매개변수 O, 반환값 O

# def 함수이름(매개변수):
#     수행할 문장
#     ...
#     return 반환값

def add(a, b): 
    result = a + b
    return result

a = add(3, 4)
print(a)
```

```python
# 2. 입력값이 없는 함수
## 매개변수 X, 반환값 O

def say():
    return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
## 매개변수 O, 반환값은 X

def add(a, b): 
    print(f"{a}, {b}의 합은 {a+b}")
    # return None

a = add(3, 4)
print(a)
```

```python
# 4. 입력값도 없고 반환값도 없는 함수
## 매개변수 X, 반환값 X

def say():
    print("hi") 

a = say()
print(a)
```

```python
# 매개변수를 지정하여 호출하기
def sub(a, b): 
    result = a - b
    return result 

c = sub(a=7, b=3)
print(c)
```

```python
d = sub(b=7, a=3)
print(d)
```

```python
# 입력값이 몇개가 될지 모를때는 어떻게 해야할까? 

# def 함수이름(*매개변수):
#     수행할 문장..

def add_many(*args):
    result = 0
    print(args)

    for i in args:
        result = result + i # result += 1 

    return result 


c = add_many(1, 2, 3, 4, 5)
print(c)
```
